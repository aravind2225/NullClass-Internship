```python
import pandas as pd
import numpy as np
```

```python
df_od=pd.read_csv("/content/orders.csv")
df_cd=pd.read_csv("/content/customers.csv")
df_pd=pd.read_csv("/content/payments.csv")
df_sk=pd.read_csv("/content/products.csv")
```

```python
df_cd.rename(columns={'id':'customer_id'},inplace=True)
df_pd.rename(columns={'id':'payment_id'},inplace=True)
df_sk.rename(columns={'id':'sku_id'},inplace=True)
```

```python
df=pd.DataFrame(df_od\
                .merge(df_sk, how='inner', on='sku_id')\
                .merge(df_cd, how='inner', on='customer_id')\
                .merge(df_pd, how='inner', on='payment_id')
)
```

```python
df.isnull().sum()
```

```python
df['order_date']=pd.to_datetime(df['order_date'])
```

```python
df['registered_date']=pd.to_datetime(df['registered_date'])
```

```python
df.info()
```

```python
#  Question 1
df_top5_Prods=df[(df['is_valid']==1) & (df['category']=='Mobiles & Tablets') & (df['order_date'].dt.year==2022)]\
.groupby('sku_name')['qty_ordered'].sum()\
.sort_values(ascending=False).head(5).reset_index()
df_top5_Prods
```

```python
import plotly.express as px
px.bar(x=df_top5_Prods['sku_name'],
            y=df_top5_Prods['qty_ordered']
            )
```

```python
df_other_cat=pd.DataFrame(
    df[(df['is_valid']==1) & (df['order_date'].dt.year==2021)]\
    .groupby('category')['qty_ordered'].sum()\
    .sort_values(ascending=False)
)
df_other_cat
```

```python
df_other_cat_2022=pd.DataFrame(
    df[(df['is_valid']==1) & (df['order_date'].dt.year==2022)]\
    .groupby('category')['qty_ordered'].sum()\
    .sort_values(ascending=False)
)
df_other_cat_2022
```

```python
df_others_2021=df[(df['is_valid']==1) & (df['category']=='Others') & (df['order_date'].dt.year==2021)]\
    .groupby('sku_name')['qty_ordered'].sum()\
    .sort_values(ascending=False).reset_index()
df_others_2021
```

```python
df_others_2022=df[(df['is_valid']==1) & (df['category']=='Others') & (df['order_date'].dt.year==2022)]\
    .groupby('sku_name')['qty_ordered'].sum()\
    .sort_values(ascending=False).reset_index()

df_others_2022
```

```python
# Q4
df['Year']=df['order_date'].dt.year
df['Month_id']=df['order_date'].dt.month
df['Month_Name']=df['order_date'].dt.month_name()
df['Day_Name']=df['order_date'].dt.day_name()
df
```

```python
req_df=pd.DataFrame(
    df[(df['Year']==2022) & (df['Month_id']>=10) & (df['is_valid']==1)]
)
```

```python
req_df.groupby('Day_Name')['before_discount'].mean().round(2).sort_values(ascending=False)
```

```python
req_df['Day_Type']=np.where(req_df['Day_Name'].isin(['Saturday','Sunday']),'WeekEnd', 'WeekDay')
req_df.groupby('Day_Type')['before_discount'].mean().round(2)
```
